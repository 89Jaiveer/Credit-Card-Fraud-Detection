```python
# 01_EDA.ipynb
# Credit Card Fraud Detection - Exploratory Data Analysis

# --------------------------------------------
# 1️⃣ Import Libraries
# --------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

```python

# --------------------------------------------
# 2️⃣ Load Dataset
# --------------------------------------------
df = pd.read_csv('/Users/jaiveersinghkhanuja/Desktop/credit_card_detection_project/creditcard.csv')
df.head()

```

```python
df.shape
df.info()
df.describe()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Normal, 1 = Fraud)")
plt.show()
```

```python
fraud_percentage = df['Class'].value_counts(normalize=True) * 100
fraud_percentage
#The dataset is extremely imbalanced. Fraud cases represent less than 1% of all transactions. Accuracy will not be a reliable metric.
```

```python
df.isnull().sum().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Amount'], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Amount vs Fraud")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
df.corr()['Class'].sort_values(ascending=False)
```

```python
top_corr = df.corr()['Class'].abs().sort_values(ascending=False)[1:6].index

for col in top_corr:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Class', y=col, data=df)
    plt.title(f"{col} vs Fraud")
    plt.show()
```
